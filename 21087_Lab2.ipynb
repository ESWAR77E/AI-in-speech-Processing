```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

new_audio_file = 'Recording.wav'  
new_y, new_sr = librosa.load(new_audio_file)  

new_derivative_1 = np.diff(new_y)
new_derivative_1 /= np.max(np.abs(new_derivative_1))

print("Playing First Derivative Signal:")
plt.figure(figsize=(10, 5))
librosa.display.waveshow(new_derivative_1, sr=new_sr, color='red')
plt.title('FIRST DERIVATIVE')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(new_derivative_1, rate=new_sr)
```

```python
new_zero_crossing = np.where(np.diff(np.sign(new_derivative_1)))[0]
new_diff = np.diff(new_zero_crossing)
new_threshold = 1000
new_speech_regions = new_diff[new_diff > new_threshold]
new_silence_regions = new_diff[new_diff <= new_threshold]

new_avg_length_speech = np.mean(new_speech_regions)
new_avg_length_silence = np.mean(new_silence_regions)

print("Average length between consecutive zero crossings in speech regions:", new_avg_length_speech)
print("Average length between consecutive zero crossings in silence regions:", new_avg_length_silence)

plt.figure(figsize=(10, 5))
plt.plot(new_diff, label='All regions', color='blue')
plt.plot(np.arange(len(new_speech_regions)), new_speech_regions, 'ro', label='Speech regions', color='green')
plt.plot(np.arange(len(new_speech_regions), len(new_speech_regions) + len(new_silence_regions)), new_silence_regions, 'bo', label='Silence regions', color='green')
plt.title('Pattern of Zero Crossings')
plt.xlabel('Zero Crossing Difference')
plt.ylabel('Difference between Consecutive Zero Crossings')
plt.legend()
plt.show()

print("Pattern of Zero Crossings:")
print("All regions:", new_diff)
print("Speech regions:", new_speech_regions)
print("Silence regions:", new_silence_regions)
```

```python

word_files_mine = ['laptop1.mp3', 'charger1.mp3', 'keyboard1.mp3', 'bag1.mp3', 'mobile1.mp3']
word_files_teammate = ['laptop2.mp3', 'charger2.mp3', 'keyboard2.mp3', 'bag2.mp3', 'mobile2.mp3']
words = ['laptop', 'charger', 'keyboard', 'bag', 'mobile']
word_lengths_mine = []
word_lengths_teammate = []

for word_file in word_files_mine:
    signal, sr = librosa.load(word_file, sr=None)
    length_seconds = len(signal) / sr
    word_lengths_mine.append(length_seconds)

for word_file in word_files_teammate:
    signal, sr = librosa.load(word_file, sr=None)
    length_seconds = len(signal) / sr
    word_lengths_teammate.append(length_seconds)

print("Lengths of the spoken words from My recordings:", word_lengths_mine)
print("Lengths of the spoken words from TeamMate's recordings:", word_lengths_teammate)

bar_width = 0.35
index = np.arange(len(words))
plt.figure(figsize=(12, 6))
plt.bar(index - bar_width/2, word_lengths_mine, bar_width, label='My Words', color='black')
plt.bar(index + bar_width/2, word_lengths_teammate, bar_width, label="Teammate's Words", color='red')
plt.xlabel('Words')
plt.ylabel('Length (seconds)')
plt.title('Comparison of Spoken Words Length')
plt.xticks(index, words)
plt.legend()

plt.show()
```
